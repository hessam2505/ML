```python
athlete.head()
```

```python
athlete['Height']=athlete['Height'].interpolate()
```

```python
athlete.groupby(['Sport'])['Height'].mean()
```

```python
athlete.groupby(['Sport','Sex']).agg({'Height':'mean','Weight':'mean'})
```

```python
pivot = athlete.pivot_table(
    columns='Sport',
    index='Sex',
    values='Height',
    aggfunc='mean'   # or 'max', 'min', 'count'
)
pivot.head()
```

```python
print('pivot : for unique values , example now we can`t use becuse we have alot of athletes that are in on sex and sport\npivot_table : you can use for every df with duplicates values and you can use mean,sum,..')
```
