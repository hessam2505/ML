### Statistics

```python
stats=np.array([[1,2,3],
                [4,5,6]])
print(stats)
print(stats.shape)
```

```python
print(np.min(stats,axis=1))

print('No Axis : search in all list')
print('Axis = 0 : search in Rows')
print('Axis = 1 : search in Columns')
```

```python
np.max(stats)
```

```python
print(np.sum(stats))
print(np.sum(stats[0]))
print(np.sum(stats[1]))
print(np.sum(stats,axis=0))
print(np.sum(stats,axis=1))
```

### Reorganizing Arrays

```python
before = np.array([[1,2,3,4],
                [5,6,7,8]])
print(before)

after= before.reshape((8,1))
print(after)
```

### Vertically stacking vectors

```python
v1=np.array([1,2,3,4])
v2=np.array([5,6,7,8])

print(np.vstack([v1,v2]))
print(np.vstack([v1,v2,v1,v2]))
```

### Horizontal stacking vactors

#### for matrix and vector

```python
h1=np.ones((2,4))
h2=np.zeros((2,2))

print(np.hstack((h1,h2)))
```

#### for vector

```python
c1=np.array([1,2,3,4])
c2=np.array([5,6,7])

print(np.concatenate((c1,c2)))
```
